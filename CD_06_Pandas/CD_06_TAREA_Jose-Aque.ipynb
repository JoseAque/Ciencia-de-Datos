<div class="alert alert-success">

<b>EJERCICIO 1 </b>: 
    
    
Encontrar el promedio de edad para los pasajeros de cada clase.
    
    
</div>

````python
promedio_C = df[df['Embarked']=='C']['Age'].mean()
promedio_Q = df[df['Embarked']=='Q']['Age'].mean()
promedio_S = df[df['Embarked']=='S']['Age'].mean()

print('Pasajeros punto C: ', promedio_C)
print('Pasajeros punto Q: ', promedio_Q)
print('Pasajeros punto S: ', promedio_S)
````

<div class="alert alert-success">

<b>EJERCICIO 2</b>: 
  
Encontrar el número total de pasajeros por cada punto de embarcación.     
    
</div>

````python
total_C = df[df['Embarked']=='C'].shape[0]
total_Q = df[df['Embarked']=='Q'].shape[0]
total_S = df[df['Embarked']=='S'].shape[0]

print('Pasajeros punto C: ', total_C)
print('Pasajeros punto Q: ', total_Q)
print('Pasajeros punto S: ', total_S)
````

<div class="alert alert-success">

<b>EJERCICIO 3</b>: 
  
a) Encontrar el promedio de edad de los pasajeros que sobrevivieron. 
    
b) Encontrar el promedio de edad de los hombres que sobrevivieron
    
c) Encontrar el promedio de edad de las mujeres que sobrevivieron
</div>


#### a) Promedio de edad de sobrevivientes

````python
print(f"El promedio de edad de los pasajeros que sobrevivieron:  {df[df['Survived']==1]['Age'].mean()}")
````

#### b) Promedio de edad de hombres sobrevivientes

````python
promedio_hombres_s = df[(df['Survived']==1) & (df['Sex']=='male')]['Age'].mean()
print(f"El promedio de edad de hombres sobrevivientes: {promedio_hombres_s}")
````

#### c) Promedio de edad de mujeres sobrevivientes

````python
promedio_mujeres_s = df[(df['Survived']==1) & (df['Sex']=='female')]['Age'].mean()
print(f"El Promedio de edad de mujeres sobrevivientes: {promedio_mujeres_s}")
````

<div class="alert alert-success">

<b>EJERCICIO 4</b>: 
  
¿Cuál fue el número de sobrevivientes mujeres  mayores a 40 años? 

</div>





````python
mujeres_40 = df[(df['Survived']==1) & (df['Sex']=='female') & (df['Age'] > 40)].shape[0]
print(f"El número de sobrevivientes mujeres mayores a 40 años: {mujeres_40}")
````

  <li> Encuentre el número de estudiantes por nivel educativo (PRIMARIA, SECUNDARIA y BACHILLERATO). </li>

</div>

````python
total_estudiantes_PRIMARIA = df[df['NIVEL']=='PRIMARIA']["MATRICULA TOTAL"].sum()
total_estudiantes_SECUNDARIA = df[df['NIVEL']=='SECUNDARIA']["MATRICULA TOTAL"].sum()
total_estudiantes_BACHILLERATO = df[df['NIVEL']=='BACHILLERATO']["MATRICULA TOTAL"].sum()

print("Total estudiantes PRIMARIA:", total_estudiantes_PRIMARIA)
print("Total estudiantes SECUNDARIA:", total_estudiantes_SECUNDARIA)
print("Total estudiantes BACHILLERATO:", total_estudiantes_BACHILLERATO)
````

<div class="alert alert-success">
    <b>EJERCICIO 7 </b>:
 <ul>
  <li>  Encuentre el número de estudiantes hombres y mujeres por nivel educativo (PRIMARIA, SECUNDARIA y BACHILLERATO).    </li>
</ul> 

</div>

````python
total_PRIMARIA_H = df[df['NIVEL']=='PRIMARIA']["HOMBRES TOTAL"].sum()
total_PRIMARIA_M = df[df['NIVEL']=='PRIMARIA']["MUJERES TOTAL"].sum()

total_SECUNDARIA_H = df[df['NIVEL']=='SECUNDARIA']["HOMBRES TOTAL"].sum()
total_SECUNDARIA_M = df[df['NIVEL']=='SECUNDARIA']["MUJERES TOTAL"].sum()

total_BACHILLERATO_H = df[df['NIVEL']=='BACHILLERATO']["HOMBRES TOTAL"].sum()
total_BACHILLERATO_M = df[df['NIVEL']=='BACHILLERATO']["MUJERES TOTAL"].sum()

print("PRIMARIA - Hombres:", total_PRIMARIA_H, "Mujeres:", total_PRIMARIA_M)
print("SECUNDARIA - Hombres:", total_SECUNDARIA_H, "Mujeres:", total_SECUNDARIA_M)
print("BACHILLERATO - Hombres:", total_BACHILLERATO_H, "Mujeres:", total_BACHILLERATO_M)

````
