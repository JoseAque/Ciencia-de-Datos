<b>EJERCICIO 1</b>: 
    
Explicar las opciones Inner,Left, Right y Outer para la función *merge*, y muestre 4 ejemplos de su uso
    
    
</div>

```python
# DataFrame 1
df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Nombre': ['Ana', 'Luis', 'Carmen', 'Jorge']
})
df1
```

```python
# DataFrame 2
df2 = pd.DataFrame({
    'ID': [3, 4, 5, 6],
    'Edad': [23, 34, 45, 29]
})
df2
```

```python
# Inner (intersección):
# Devuelve solo las filas que tienen coincidencia en ambos DataFrames.
# → "Lo que tienen en común".

inner = pd.merge(df1, df2, on='ID', how='inner')
print(inner)
```

```python
# Left (izquierda):
# Devuelve todas las filas del DataFrame izquierdo, y añade valores del derecho cuando hay coincidencia.
# Si no hay coincidencia, rellena con NaN.
# → "Todo el izquierdo, lo que coincida del derecho".

left = pd.merge(df1, df2, on='ID', how='left')
print(left)
```

```python
# Right (derecha):
# Lo opuesto al Left. Devuelve todas las filas del derecho y añade valores del izquierdo cuando hay coincidencia.
# → "Todo el derecho, lo que coincida del izquierdo".

right = pd.merge(df1, df2, on='ID', how='right')
print(right)
```

```python
# Outer (unión completa):
# Devuelve todas las filas de ambos DataFrames, coincidan o no.
# Cuando no hay coincidencia, rellena con NaN.
# → "La unión total".

outer = pd.merge(df1, df2, on='ID', how='outer')
print(outer)
```
